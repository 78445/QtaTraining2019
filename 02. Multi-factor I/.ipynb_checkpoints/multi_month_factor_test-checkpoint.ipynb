```python
#该脚步需要在JointQuant上运行

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

factorNameList = ['market_cap', 'ROE']
```

```python
def normalization(aSeries):
    sMean = aSeries.mean()
    sStd = aSeries.std()
    aSeries = (aSeries-sMean)/sStd
    return(aSeries)
```

```python
def getGroupReturn(group, startdate, enddate):
    aGroupDict = pd.DataFrame()
    for aIndex in group.index:        
        try:
            aGroupDict[aIndex] =get_price(aIndex, start_date=startdate, end_date=enddate,panel=False,fields = ['close'])['close']
#             print(get_price(aIndex, start_date=startdate, end_date=enddate,panel=False,fields = ['close']))
        except :
            print(aIndex)
        
    aGroupReturn = aGroupDict.pct_change().mean(axis = 1)
    #print(aGroupReturn)
    return(aGroupReturn)

    
```

```python
def singleFactorTest(factorName,neutral=True):
    #一些数据预定义
    numberOfGroup = 5
    totalGroupList = []
    
    #股票池为上证100指数的所有股票
    stock_set = get_index_stocks('000001.XSHG',"2018-01-01")
    #查询内容需要和factorNameList对应
    q = query(
        valuation.code,
        valuation.circulating_market_cap, 
        income.net_profit/balance.total_owner_equities
    ).filter(
        valuation.code.in_(stock_set)
    )
    
#     init
    for i in range(numberOfGroup):
        totalGroupList.append(pd.Series())
    
    #声明时间窗口，月度调仓
    timerange = pd.date_range("2015-01","2018-06",freq="1M")
    for change_hand_date in timerange:#[:-1]:
        if(change_hand_date==timerange[len(timerange)-1]):
            break
        ##数据获取与预处理
        
        #获取指定日期的数据
        df_for_hand_change = get_fundamentals(q, change_hand_date)
        
#         print(df_for_hand_change)
        
        df_for_hand_change.columns = ['code']+factorNameList
        
        #获取指定因子，构建dataFrame
        df_factors = df_for_hand_change.set_index("code").loc[:,factorNameList]
        
        #正规化
        df_factors = df_factors.apply(normalization)
        
        #补充NaN缺失值
        df_factors = df_factors.fillna(method = 'ffill')
        
        #通过回归获得与其他因子中性的残差项代替
        #print(df_factors)
        if len(factorNameList)>1 and neutral :
            X_train = np.array(df_factors.drop(factorName,axis = 1))
            Y_train = np.array(df_factors.loc[:,factorName])
            Y_train.reshape(1,-1)
            linreg = LinearRegression()

            try:
                linreg.fit(X_train, Y_train)
            except:
                print(change_hand_date)
                print(X_train)
            linreg.score(X_train, Y_train)
            risidual = Y_train - linreg.predict(X_train)
            df_factors[factorName+'risidual'] =risidual
        else:
            df_factors[factorName+'risidual']=np.array(df_factors.loc[:,factorName])
        
        
        ##股票分组与检测
        
        df_sorted_by_single_factor = df_factors.sort_values(factorName+'risidual', ascending = False)
        sortedSecurities = get_all_securities().loc[df_sorted_by_single_factor.index]
        numInGroup = int(len(sortedSecurities)/numberOfGroup)
        
        groupList = []
        for i in range(numberOfGroup):
            if i == numberOfGroup-1:
                groupList.append(sortedSecurities[i*numInGroup:])
            else:
                groupList.append(sortedSecurities[i*numInGroup:(i+1)*numInGroup])
        
        for aGroup_i in range(numberOfGroup):
            
            aGroupReturn = getGroupReturn(groupList[aGroup_i], startdate = change_hand_date, enddate = timerange[timerange.get_loc(change_hand_date)+1])
            
            totalGroupList[aGroup_i] = pd.concat([totalGroupList[aGroup_i],aGroupReturn])
    
        
    #添加基准收益率以进行比较
    benchmark = get_price(['000001.XSHG'],start_date=timerange[0],end_date=timerange[-1],panel=False).set_index('time')['close']
    benchmark = benchmark.pct_change()
    totalGroupList.append(benchmark)
    #print(benchmark)
    #画图展示
    plt.figure(figsize=(30,12))
    local_handles = []
    for num ,aGroup in enumerate(totalGroupList):
        aGroup=aGroup.fillna(method = 'ffill')
        line_data = (aGroup+1).cumprod()-1
#         print(num, line_data)
        hd, = plt.plot(line_data)
        local_handles.append(hd)


    plt.legend(handles=local_handles, labels=range(numberOfGroup+1),
    loc='best',fontsize=30)
    plt.title(factorName,fontsize=30)
    plt.show()
        
        
```

```python
for name in factorNameList:
    singleFactorTest(name,neutral=True)
```
